```python
from transformers import pipeline
```

```python
gpt2_medium_qa_pipeline = pipeline('text-generation', model='gpt2-medium')
```

```python
def clean_generated_answer(answer):
    text = answer[0]['generated_text'].strip()
    cleaned_text = text.replace('\n\n', ' ').replace('\n', ' ')
    return cleaned_text
```

```python
# Define a function to generate answers using the pipeline
def generate_answer_with_pipeline(question: str, pipeline):
    prompt = question
    output = pipeline(prompt, max_new_tokens=200, top_k=4, pad_token_id=50256, return_full_text=False)
    clean_answer = clean_generated_answer(output)
    return clean_answer
```

```python
# Example usage
question = "What is the difference between concatenation vs. summation of two tensors?"
answer = generate_answer_with_pipeline(question, gpt2_medium_qa_pipeline)
print("Generated Answer:", answer)
```
